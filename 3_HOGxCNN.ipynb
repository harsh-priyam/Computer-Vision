```python
import dlib
import cv2
```

```python
image = cv2.imread('/content/drive/MyDrive/Images/people2.jpg')
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(image)
```

```python
face_detector_hog = dlib.get_frontal_face_detector()
```

```python
detection = face_detector_hog(image,1)
```

```python
detection
```

```python
for face in detection:
  l,t,r,b = face.left(),face.top(),face.right(),face.bottom()
  cv2.rectangle(image,(l,t),(r,b),(0,255,0),1)

cv2_imshow(image)
```

```python
# FACE RECOGNITION WITH CNN & DLIB
```

```python
image = cv2.imread('/content/drive/MyDrive/Images/people2.jpg')
```

```python
cnn_detector = dlib.cnn_face_detection_model_v1('/content/drive/MyDrive/Weights/mmod_human_face_detector.dat')
```

```python
detections = cnn_detector(image,1)
for face in detections:
  l,t,r,b,c = face.rect.left(),face.rect.top(),face.rect.right(),face.rect.bottom(),face.confidence
  print(c)
  cv2.rectangle(image,(l,t),(r,b),(255,255,0),2)
cv2_imshow(image)  
```

```python
image2 = cv2.imread('/content/drive/MyDrive/Images/people3.jpg')
face_detector_hog = dlib.get_frontal_face_detector()
detections = face_detector_hog(image2,4)
for face in detections:
  l,t,r,b = (face.left(),face.top(),face.right(),face.bottom())
  cv2.rectangle(image2,(l,t),(r,b),(255,0,0),1)
cv2_imshow(image2)  
```
